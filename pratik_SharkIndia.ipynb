#### Now, i want to plot a bar in which, i want the data of each shark that how many time they did a deal (Y) with brand 

```python
Anu = a1[a1.Anupam=='Y'].shape[0]
Ash = a1[a1.Ashneer=='Y'].shape[0]
Nam = a1[a1.Namita=='Y'].shape[0]
Am = a1[a1.Aman=='Y'].shape[0]
Pey = a1[a1.Peyush=='Y'].shape[0]
Vin = a1[a1.Vineeta=='Y'].shape[0]
Gha = a1[a1.Ghazal=='Y'].shape[0]
x = [Anu,Ash,Nam,Am,Pey,Vin,Gha]
x
```
